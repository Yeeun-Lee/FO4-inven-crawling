```
# 어절 n-gram
def seg_ngram(comments, max_num):
    n_gram = {}
    for num in range(2, max_num+1): #num : gram의 개수(2~max, minimun=bigram)
        print("Current {}-Gram".format(num))
        for text in tqdm(comments):
            text = text.replace(' ', '')
            #text = text.replace('토티', 'TOTY')
            for x in range(len(text)):
                temp = ''.join(text[x:x+num])
                if temp in n_gram: #n_gram에 이미 존재하는 경우
                    n_gram[temp]+=1
                else:#n_gram에 없는 경우
                    n_gram[temp] = 1
    scores = n_gram.items()
    scores = sorted(scores, key=lambda x:x[1], reverse=True) #많은 수대로 출력
    return scores
```

```
n_grams = seg_ngram(comments, 5)
```
